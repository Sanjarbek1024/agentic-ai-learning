````python
from sentence_transformers import SentenceTransformer
````

````python
model_name = 'sentence-transformers/all-MiniLM-L6-v2'

model = SentenceTransformer(model_name)
````

````python
model
````

````python
sentence1 = 'I love you.'
sentence2 = 'I like you.'
sentence3 = 'I hate you.'
````

````python
result1 = model.encode(sentence1)
result2 = model.encode(sentence2)
result3 = model.encode(sentence3)
````

````python
from sentence_transformers import util

# Calculate cosine similarity matrix for all three sentences
similarities = util.cos_sim([result1, result2, result3], [result1, result2, result3])

print("Similarity matrix:\n", similarities)
````

````python
import numpy as np

# Similarity between 'I love you' and 'I like you'
sim_1_2 = np.dot(result1, result2)

# Convert similarity to Cosine Distance (0 means identical, 2 means opposites)
dist_1_2 = 1 - sim_1_2

print(f"Cosine Distance: {dist_1_2}")
````

````python
import torch

def mean_pooling(model_output, attention_mask):
    token_embeddings = model_output[0] #First element of model_output contains all token embeddings
    input_mask_expanded = attention_mask.unsqueeze(-1).expand(token_embeddings.size()).float()
    return torch.sum(token_embeddings * input_mask_expanded, 1) / torch.clamp(input_mask_expanded.sum(1), min=1e-9)
````

````python
documents = [
    "Python is a popular programming language.",
    "Football is the world's most popular sport.",
    "Artificial intelligence is changing many industries.",
    "Cats are friendly domestic animals.",
    "Machine learning is a subset of artificial intelligence.",
    "Messi is one of the greatest football players.",
    "Pandas is a Python library for data analysis."
]
````

````python
document_embeddings = model.encode(documents)
````

### Bu odatda bir marta qilinadi va Vector DB (FAISS, Chroma va boshqalar)ga saqlanadi.

````python
query = "How can I learn Python?"
````

````python
query_embedding = model.encode(query)
````

````python
from sklearn.metrics.pairwise import cosine_similarity

scores = cosine_similarity(
    [query_embedding],
    document_embeddings
)[0]
````

````python
scores
````

````python
import numpy as np

top_k = 3

top_indices = np.argsort(scores)[::-1][:top_k]

print(query)
print("-" * 40)

for idx in top_indices:
    print(f"Score: {scores[idx]:.4f}")
    print(documents[idx])
    print("-" * 40)
````
