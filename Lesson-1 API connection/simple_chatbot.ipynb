```python
import os
from google import genai
from dotenv import load_dotenv
```

```python
load_dotenv()
```

```python
model_name = 'gemini-3.5-flash'
client = genai.Client()
```

```python
print('This chatbot is running. So ask anything you want!')
user_text = str(input("Write your question: "))


while True:
    output = client.interactions.create(
        model=model_name,
        input=user_text
    )
    
    print("\nQuestion: ", user_text)
    print('Answer: ', output.output_text)
    
    asking = str(input('Do you want to continue (Yes, No): '))
    
    if asking.lower().strip() == 'yes':
        user_text = str(input("Write your question: "))
        continue
    else:
        break
```
